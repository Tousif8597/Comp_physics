```python
def Bisect_Maximum(f, a, b, tol):
    """This is a method for maximum in between two points.
    Parameters: 
    f = function of which maximum will be found.
    a = lower end point.
    b = higher end pont.
    tol = error.
    xmax = point of maximum
    f(xmax) = maximum value of the function
    """
    
    while abs(b - a) > tol:
        c = a + (b - a)/3
        d = b - (b - a)/3
        if f(c) > f(d):  
            b = d  #b is updated to d so in the next loop we will find in [a, d] 
                   #since from a to c function increases comparably faster so in [c, b] it may not contain the maximum
        else:
            a = c
    xmax = (a + b)/2
    return xmax, f(xmax)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x): return np.sin(3*x) + np.cos(2*x)
x = np.linspace(-3, 3, 5000)
plt.plot(x, f(x))
plt.axvline(0, linestyle = '--', color = 'crimson', label = "Lower end[a]")
plt.axvline(1, linestyle = '-.', color = 'crimson', label = "Higher end[b]")
plt.grid()
plt.legend()
plt.title("f(x) = sin(3x) + cos(2x) ")
plt.show()
```

```python
xm, f_xm = Bisect_Maximum(f, 0, 1, 1e-8)
```

```python
print("Point of maximum =",xm)
print("Maximum value =",f_xm)
```
